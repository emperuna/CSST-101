```python
# ============================================================
# Computer Troubleshooting Assistant
# Author: Garin, Jeremy M.
# Section: BSCS - 3A
# Description:
#   A rule-based expert system that provides troubleshooting
#   advice for common computer problems using conditional rules.
# ============================================================

def troubleshoot():
    """Rule-Based Expert System for Computer Troubleshooting."""

    print("===============================================")
    print("     Welcome to the Computer Troubleshooting Assistant")
    print("===============================================")
    print("Type 'exit' anytime to close the program.\n")

    # Continuous loop until user types 'exit'
    while True:
        # Get user input
        problem = input("Describe your computer problem: ").lower()

        # Exit condition
        if problem == "exit":
            print("\nExiting the Troubleshooting Assistant. Goodbye!")
            break

        # Rule 1: No Power
        if "no power" in problem or "won't turn on" in problem:
            print("\nCondition: Computer has no power or won't turn on.")
            print("Conclusion: Possible Solutions:")
            print("- Check if the power cord is properly plugged in.")
            print("- Try another outlet or power source.")
            print("- If laptop, ensure the battery or charger works.")
            print("- Check the PSU if possible.")

        # Rule 2: No Display (Chained Rule Example)
        elif "no display" in problem or "blank screen" in problem:
            print("\nCondition: Computer powers on but shows no display.")
            print("Conclusion: Possible Solutions:")
            print("- Ensure monitor is turned on and cables are secure.")
            print("- Try a different video cable or monitor.")
            print("- Ensure the monitor is connected to the graphics card port.")

            # ðŸ”— Chained Rule
            response = input("\nDo you hear any beeping sounds when turning on the computer? (yes/no): ").lower()
            if response == "yes":
                print("\nâ†’ The beeps indicate a hardware issue (RAM or GPU). Reseat or test components.")
            else:
                print("\nâ†’ No beeps may indicate a motherboard or power delivery issue.")

        # Rule 3: Slow Performance
        elif "slow performance" in problem or "computer is slow" in problem:
            print("\nCondition: Computer is running slowly.")
            print("Conclusion: Possible Solutions:")
            print("- Close unused background applications.")
            print("- Run a virus/malware scan.")
            print("- Free up disk space or upgrade to SSD.")
            print("- Add more RAM if possible.")

        # Rule 4: No Internet Connection
        elif "internet not working" in problem or "no internet" in problem:
            print("\nCondition: Internet connection not working.")
            print("Conclusion: Possible Solutions:")
            print("- Check if Wi-Fi or Ethernet is properly connected.")
            print("- Restart your modem or router.")
            print("- Check your network adapter settings.")
            print("- Contact your Internet Service Provider (ISP).")

        # Rule 5: Keyboard/Mouse Not Working
        elif "keyboard not working" in problem or "mouse not working" in problem:
            print("\nCondition: Keyboard or mouse is not responding.")
            print("Conclusion: Possible Solutions:")
            print("- Ensure the device is properly connected.")
            print("- For wireless devices, replace batteries.")
            print("- Try a different USB port or reconnect the receiver.")
            print("- Update or reinstall drivers.")

        # Rule 6: Loud Noise or Fan Issue
        elif "computer making noise" in problem or "fan is loud" in problem:
            print("\nCondition: Computer is making unusual noise.")
            print("Conclusion: Possible Solutions:")
            print("- Clean dust from fans and vents.")
            print("- Ensure proper airflow and ventilation.")
            print("- Clicking sounds may indicate a failing hard drive.")
            print("- Check for loose components inside the case.")

        # Default Rule
        else:
            print("\nCondition: Unknown or unlisted issue.")
            print("Conclusion: Please consult a computer technician for further help.")

        print("\n-----------------------------------------------\n")

# Run the system
if __name__ == "__main__":
    troubleshoot()
```
